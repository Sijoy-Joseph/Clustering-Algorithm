```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
X = df
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
kmeans_clusters = kmeans.predict(X)
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans_clusters, cmap='viridis')
plt.title("KMeans Clustering (2D Visualization)")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
#Brief description of how Hierarchical clustering works:

Hierarchical clustering builds a hierarchy of clusters by either starting with individual points and merging them (agglomerative) or starting with all points in one cluster and recursively dividing them (divisive).
In agglomerative hierarchical clustering:
Each point is considered a cluster.
At each step, the two closest clusters are merged.
The process continues until only one cluster remains.
The result of hierarchical clustering is often represented by a dendrogram, which shows the arrangement of the clusters at each level.
Why Hierarchical clustering might be suitable for the Iris dataset:

Hierarchical clustering is beneficial when the number of clusters is not known beforehand (though we know the Iris dataset has 3 clusters, it can still reveal interesting patterns at different levels).
It is good for understanding the structure and hierarchy of the data.
```
